```python
import os
import cv2
import numpy as np
import random

# Function to add blur defect to image
def add_blur_defect(image):
    blurred = cv2.GaussianBlur(image, (21, 21), 0)
    return blurred

# Function to add point defect to image
def add_point_defect(image):
    x, y = random.randint(0, image.shape[1] - 1), random.randint(0, image.shape[0] - 1)
    cv2.circle(image, (x, y), 5, (0, 0, 0), -1)
    return image

# Define function to add square defect to image
def add_square_defect(image):
    x, y = random.randint(0, image.shape[1] - 50), random.randint(0, image.shape[0] - 50)
    cv2.rectangle(image, (x, y), (x + 50, y + 50), (0, 0, 0), -1)
    return image

# Function to add line defect to image
def add_line_defect(image):
    x1, y1 = random.randint(0, image.shape[1]), random.randint(0, image.shape[0])
    x2, y2 = random.randint(0, image.shape[1]), random.randint(0, image.shape[0])
    cv2.line(image, (x1, y1), (x2, y2), (0, 0, 0), 5)
    return image

# Function to add triangular defect to image
def add_triangular_defect(image):
    x1, y1 = random.randint(0, image.shape[1]), random.randint(0, image.shape[0])
    x2, y2 = random.randint(0, image.shape[1]), random.randint(0, image.shape[0])
    x3, y3 = random.randint(0, image.shape[1]), random.randint(0, image.shape[0])
    points = np.array([[x1, y1], [x2, y2], [x3, y3]], np.int32)
    cv2.fillPoly(image, [points], (0, 0, 0))
    return image

# Function to add circular defect to image
def add_circular_defect(image):
    x, y = random.randint(0, image.shape[1]), random.randint(0, image.shape[0])
    cv2.circle(image, (x, y), 20, (0, 0, 0), -1)
    return image

# Create directory for faulted dataset if it doesn't exist
faulted_dir = '/content/gdrive/MyDrive/BTP2_/dataset_faulted'
os.makedirs(faulted_dir, exist_ok=True)

# Process each image in the dataset
for i in range(1, 1015):
    # Load original image
    original_image_path = f'/content/gdrive/MyDrive/BTP2_/dataset/Image_{i}.jpg'
    original_image = cv2.imread(original_image_path)

    # Randomly choose a defect type
    defect_type = random.choice(['blur', 'point', 'square', 'line', 'triangular', 'circular'])

    # Apply defect based on chosen type
    if defect_type == 'blur':
        faulty_image = add_blur_defect(original_image)
    elif defect_type == 'point':
        faulty_image = add_point_defect(original_image)
    elif defect_type == 'square':
        faulty_image = add_square_defect(original_image)
    elif defect_type == 'line':
        faulty_image = add_line_defect(original_image)
    elif defect_type == 'triangular':
        faulty_image = add_triangular_defect(original_image)
    elif defect_type == 'circular':
        faulty_image = add_circular_defect(original_image)

    # Save faulty image to faulted dataset folder
    faulty_image_path = f'{faulted_dir}/Image_{i}_defected.jpg'
    cv2.imwrite(faulty_image_path, faulty_image)

    print(f'Created faulty image: {faulty_image_path}')

```
